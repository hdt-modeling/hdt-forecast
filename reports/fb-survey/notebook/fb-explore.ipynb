## The number of missing values per question

Let's look at the number of missing values in each question. Keep in mind that the missingness can be due to survey design logic and the changes made over time for different waves.

```python
# Count the missing values for each column in the SQL table
# Construct the query
query = """SELECT COUNT(CASE WHEN %s IS NULL OR %s="NaN" then 1 end) as %s,"""%(questions.columns[0],questions.columns[0],questions.columns[0])
for question in questions.columns[1:]:
    # If it is not the last element, we replicate the statement without the tablename
    if question is not questions.columns[-1]:
        query  = query + " " + """COUNT(CASE WHEN %s IS NULL OR %s="NaN" then 1 end) as %s,"""%(question,question,question)
    # if it is, we append the tablename at the end
    else:
        query  = query + " " + """COUNT(CASE WHEN %s IS NULL OR %s="NaN" then 1 end) as %s from survey"""%(question,question,question)
# Run the query
num_missing_total_df = pd.read_sql_query(query, engine)
# show the transpose to get a tall dataframe instead
t_num_missing_total_df = num_missing_total_df.T
t_num_missing_total_df["questions"] = questions.columns
t_num_missing_total_df.columns = ["Count", "questions"]
```

```python
dodge_text = p9.position_dodge(width=0.9)

(p9.ggplot(t_num_missing_total_df, p9.aes(x='reorder(questions, Count)', y='Count', fill='Count')) 
 + p9.geom_bar(stat = "identity", position='dodge')
 + p9.coord_flip()
 + p9.scale_fill_gradient(low='#56B1F7', high='#132B43')
 + p9.labs(y='Count', x='Question Code', title='The total number of missing values per question')
 + p9.theme(figure_size=(8, 16))
)
```

## Percentage of missingness by wave

Instead of looking at the total number of missing values, we may want to look at missing values by wave to understand better the missing values distribution. It is because the questions are controlled by waves. 

```python
# Count the missing values for each column in the SQL table
# Construct the query
query = """SELECT COUNT(CASE WHEN %s IS NULL OR %s="NaN" then 1 end) as %s,"""%(questions.columns[0],questions.columns[0],questions.columns[0])
for question in questions.columns[1:]:
    # If it is not the last element, we replicate the statement without the tablename
    if question is not questions.columns[-1]:
        query  = query + " " + """COUNT(CASE WHEN %s IS NULL OR %s="NaN" then 1 end) as %s,"""%(question,question,question)
    # if it is, we append the tablename at the end
    else:
        query  = query + " " + """COUNT(CASE WHEN %s IS NULL OR %s="NaN" then 1 end) as %s from survey GROUP BY wave"""%(question,question,question)
# Run the query
view = pd.read_sql_query(query, engine)
view
```

```python
# Join with the response by wave
joined_missing = pd.concat([view,response_by_wave], axis = 1)
# Compute the percentages
missing_percent = joined_missing.iloc[:,:(joined_missing.shape[1]-2)].div(joined_missing.num_of_responses, axis=0)*100.0
response_rate_by_wave = 100-missing_percent
response_rate_by_wave = pd.concat([response_rate_by_wave,response_by_wave], axis = 1)
response_rate_by_wave
```

```python
col = response_rate_by_wave.columns[:98]
x = np.array(["1","2","3","4","5","6","7"])
new_df = pd.concat([response_rate_by_wave.iloc[0,:98],response_rate_by_wave.iloc[1,:98],response_rate_by_wave.iloc[2,:98],response_rate_by_wave.iloc[3,:98],response_rate_by_wave.iloc[4,:98],response_rate_by_wave.iloc[5,:98],response_rate_by_wave.iloc[6,:98]])
new_df = new_df.to_frame()
new_df["Questions"]= col.to_list()+col.to_list()+col.to_list()+col.to_list()+col.to_list()+col.to_list()+col.to_list()
new_df["wave"]= np.repeat(x, [98,98,98,98,98,98,98], axis=0).tolist()
new_df.columns = ["Percentage", "Questions", "Wave"]
```

```python
# Visualize the response rate by wave for each question
(p9.ggplot(new_df, p9.aes(x='Questions', y='Percentage', fill='Wave'))
 + p9.coord_flip()
 + p9.geom_col(stat='identity', position='dodge')
 + p9.theme(figure_size=(10, 55))
)  
```

### Let's look at the graphs above by using percentage of response by state

We should account for the number of response while looking at the trends. We can visualize the percentage of the response instead to see if the trend changes. 


We can see that percentage of saying definitely no or probably no to getting vaccinated out of all those responsed the questions in CA is decreasing over time.

```python
query = """
    SELECT t.Time, t.state_name, n.num_of_response / t.num_of_response * 100.0 as Percentage_of_No
    FROM(SELECT s.Time, a.state_name, s.num_of_response
        FROM (SELECT EndDatetime as Time, A3b as State, CAST(COUNT(*) AS float) as num_of_response
        FROM survey 
        WHERE wave >= 6 and (V3 == "3.0" or V3 == "4.0")
        GROUP BY A3b, EndDatetime) s 
    INNER JOIN area a 
    WHERE s.State = a.state_id) n
    INNER JOIN
    (SELECT s.Time, a.state_name, s.num_of_response
        FROM (SELECT EndDatetime as Time, A3b as State, CAST(COUNT(*) AS float) as num_of_response
        FROM survey 
        WHERE wave >= 6 and V3 IS NOT NULL
        GROUP BY A3b, EndDatetime) s 
    INNER JOIN area a 
    WHERE s.State = a.state_id) t
    WHERE n.Time = t.Time AND n.state_name = t.state_name
"""
percent_no = pd.read_sql_query(query, engine)
```

```python
(
    p9.ggplot(percent_no, p9.aes(x='Time', y='Percentage_of_No', group = 1))
    + p9.geom_line() # line plot
    + p9.labs(x='Date', y='Percentage of no to getting vaccinated if offered', title="Percentage of no (probably+definitely) to getting vaccinated if offered out of all who responded the question")
    + p9.theme(figure_size=(12, 16), axis_text_x=p9.element_blank())
    + p9.facet_wrap('state_name', nrow=9)
)
```

```python
# Zoom in to look at California
(
    p9.ggplot(percent_no[percent_no["state_name"]=="California"], p9.aes(x='Time', y='Percentage_of_No', group = 1))
    + p9.geom_line() # line plot
    + p9.labs(x='Date', y='Percentage', title="Percentage of no (probably+definitely) to getting vaccinated if offered out of all who responded the question")
    + p9.theme(figure_size=(7, 4), axis_text_x=p9.element_text(rotation=90))
)
```

## In the past 5 days, have you traveled outside of your state?

Time period may vary from fips to fips due to the missing values. The following covers survey wave 1 (starting from April 6, 2020) to wave 7 (Jan 31, 2021).

* Yes (1)
* No (2)

```python
query = """
    SELECT t.Date, t.fips, s.num_of_response as num_yes, t.num_of_response as total_response, s.num_of_response / t.num_of_response * 100.0 as percentage_of_yes
    FROM(SELECT EndDatetime as Date, fips, CAST(COUNT(*) AS float) as num_of_response
    FROM survey
    WHERE C6 = 1.0 and fips LIKE '6%'
    GROUP BY fips, EndDatetime) s
    INNER JOIN
    (SELECT EndDatetime as Date, fips, CAST(COUNT(*) AS float) as num_of_response
    FROM survey
    WHERE C6 != "NaN" and C6 IS NOT NULL and fips LIKE '6%'
    GROUP BY fips, EndDatetime) t
    WHERE t.Date = s.Date AND t.fips = s.fips
"""
travel = pd.read_sql_query(query, engine)
travel
```

```python
(
    p9.ggplot(travel, p9.aes(x='Date', y='percentage_of_yes', group = 1))
    + p9.geom_line() # line plot
    + p9.labs(x='Date', y='Percentage of yes', title="Percentage of yes that traveled in the past 5 days within each fips in CA")
    + p9.theme(figure_size=(12, 16), axis_text_x=p9.element_blank())
    + p9.facet_wrap('fips', nrow=9)
)
```
